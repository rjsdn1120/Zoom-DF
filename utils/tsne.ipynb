```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

data = load_digits()

n_components = 2

model = TSNE(n_components=n_components)

transformed=model.fit_transform(data.data)
print(transformed)
# [
#     [67.38322, -1.9517338],
#     [-11.936052, -8.906425],
#     ...
#     [-10.278599, 8.832907],
#     [25.714725, 11.745557],
# ]
```
